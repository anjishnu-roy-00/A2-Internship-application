<h2>Function for extraction</h2>

```python
def extractor(item,url):
    name = item.find('div', {'class' :'sc-f8165ac8-17 WwCBf'}).get('title').strip()
    currency = item.find('span', {'class' :'currency'}).text.strip()
    price = item.find('strong').text
    try:
        rating = item.find('span', {'class' :'ratingValue'}).text
        rating_count = item.find('span', {'class' :'count'}).text
    except AttributeError:
        rating = ''
        rating_count = '' 
    if item.find('div', {'class' : 'sc-b823bcb8-0 jYIGEK'}) == None:
        FBN = 'Market'
    else:
        FBN = 'Express'
    URL = "https://www.noon.com"+str(url.a.get('href'))
    r = requests.get(URL)
    soup_0 = BeautifulSoup(r.content, 'html.parser', )
    Seller = soup_0.find('span', {'class' : 'allOffers'}).text
    final = {
        'Name' : name,
        'Average Rating': rating,
        'Rating Count' : rating_count,
        'Currency' : currency,
        'Price' : price,
        'FBN' : FBN,
        'URL' : URL,
        'Seller' : Seller
    }
    return final
```

<h2>Getting to next page</h2>

```python
url_fin = 'https://www.noon.com/egypt-en/sports-and-outdoors/exercise-and-fitness/yoga-16328/?limit=50&page=1&sort[by]=popularity&sort[dir]=desc'
```

```python
driver2 = webdriver.Firefox(options=options)
```

```python
def connect(link,page):
    driver2.get(link)
    time.sleep(10)
    results = []
    spans = []
    soup = BeautifulSoup(driver2.page_source, 'html.parser', )
    results = soup.find_all('div', {'class' : 'sc-f8165ac8-15 bkbUJe'})
    spans = soup.find_all('span', {'class' : 'sc-5e739f1b-0 gEERDr wrapper productContainer'})
    print(len(results))
    for u in range(len(results)):
        print("Extracting from page %s and product %s"%(page,(u+1)))
        ret = extractor(results[u],spans[u])
        records.append(ret)
```
